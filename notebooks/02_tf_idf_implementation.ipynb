### Implementação de TF-IDF
Transformação do texto em representação numérica utilizável para o modelo de IA.

```python
import pandas as pd
from pathlib import Path
from sklearn.feature_extraction.text import TfidfVectorizer

PROCESSED_PATH = Path("../data/processed/movies.parquet")
FEATURES_PATH = Path("../data/features/tfidf_matrix.joblib")
VECTORIZER_PATH = Path("../data/features/tfidf_vectorizer.joblib")
```

```python
df = pd.read_parquet(PROCESSED_PATH)

df
```

```python
vectorizer = TfidfVectorizer(
    stop_words="english",
    max_features=5000, 
    ngram_range=(1, 2)
)

tfidf_matrix = vectorizer.fit_transform(df["content"])

print(tfidf_matrix.shape)
print(len(vectorizer.get_feature_names_out()))
```

### Aplicação do Sistema de Recomendação

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity_matrix = cosine_similarity(tfidf_matrix)
```

```python
# Teste prático
idx = df[df["title"] == "The Matrix"].index[0]

scores = list(enumerate(similarity_matrix[idx]))
scores = sorted(scores, key=lambda x: x[1], reverse=True)

top_movies = scores[1:6]

top_movies # índices + scores de similaridade
```

```python
recommendations = []

for i, score in top_movies:
    title = df.iloc[i]["title"]
    recommendations.append((title, score))

print('Filmes recomendados para "The Matrix":\n')

for i, (title, score) in enumerate(recommendations, start=1):
    print(f"{i}. {title} ({round(score * 100, 1)}% similar)")
```
