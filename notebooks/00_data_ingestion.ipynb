### Busca de dados brutos

```python
from dotenv import load_dotenv
from pathlib import Path
import os
import json
import requests
import pandas as pd


load_dotenv()

API_KEY = os.getenv("API_KEY")
BASE_URL = os.getenv("BASE_URL")

OUTPUT_PATH = Path("../data/raw/movies_raw.json")

if not API_KEY:
    raise ValueError("API KEY Não foi encontrada nas variáveis de ambiente.")
```

```python
pages = 5 # busca de páginas de teste
url = f"{BASE_URL}/movie/popular"
params = {
    "api_key": API_KEY,
    "language": "en-US",
    "page": pages
}

response = requests.get(url, params=params)

if response.status_code != 200:
    raise Exception(f"Erro na API: {response.status_code}")

data = response.json()
df = pd.json_normalize(data)
```

### Etapa de ingestão

```python
all_movies = []

results = data.get("results", [])
all_movies.extend(results)

all_movies
```

### Armazenamento na pasta de dados sem tratamento

```python
OUTPUT_PATH.parent.mkdir(parents=True, exist_ok=True)

with open(OUTPUT_PATH, "w", encoding="utf-8") as f:
    json.dump(all_movies, f, ensure_ascii=False, indent=2)
```
